## Geocoding API

```python
import requests,json
```

```python
def getcoordinates():
    apikey='d26e65b197b0476feb70f95dbcc27ce1'
    city='Ahmedabad'
    URL=f'http://api.openweathermap.org/geo/1.0/direct?q={city}&appid={apikey}'
    response=requests.get(URL)
    print(response)
    coordinates=response.json()
    print(coordinates)
    latitude=coordinates[0]['lat']
    longitude=coordinates[0]['lon']
    
    return latitude,longitude
lat,long=getcoordinates()
```

## Air polution API

```python
import pprint
def getAirpollution(lat,lon):
    apikey='d26e65b197b0476feb70f95dbcc27ce1'
    pol_url=f'http://api.openweathermap.org/data/2.5/air_pollution?lat={lat}&lon={lon}&appid={apikey}'
    response=requests.get(pol_url)
    print(response)
    pollution=response.json()
    return pollution
    
pollution=getAirpollution(lat,long)
pprint.pprint(pollution)
```

```python
index=pollution['list'][0]['main']['aqi']
print('AQI - ',index)
```

```python
quality={1:'Good',2:'Fair',3:'Moderate',4:'Poor',5:'Very Poor'}
print('Quality - ',quality[index])
```

## 5 Days Forecast

- Create a dataframe of columns date,tempreture,pressure,humidity and weather discription

```python
def forecast(lat,lon):
    apikey='d26e65b197b0476feb70f95dbcc27ce1'
    url=f'http://api.openweathermap.org/data/2.5/forecast?lat={lat}&lon={lon}&appid={apikey}&lang=hi&units=metric'
    response=requests.get(url)
    forecast=response.json()
#     pprint.pprint(forecast)
    
    return forecast

data=forecast(lat,long)
print(len(data['list']))
```

```python
data
```

```python
D={'Date':[],
'Tempreture':[],
'Pressure':[],
'Humidity':[],
'Weather_Description':[]}
```

```python
for i in data['list']:
    D['Date'].append(i['dt_txt'])
    D['Tempreture'].append(i['main']['temp'])
    D['Pressure'].append(i['main']['pressure'])
    D['Humidity'].append(i['main']['humidity'])
    D['Weather_Description'].append(i['weather'][0]['description'])
```

```python
import pandas as pd
df=pd.DataFrame(D)
df
```

## NASA API

```python
import requests,json
```

```python
apikey='DEMO_KEY'
ROVER_NAME='curiosity'
SOL=1000
CAMERA='NAVCAM'
URL=f"https://api.nasa.gov/mars-photos/api/v1/rovers/{ROVER_NAME}/photos"
```

```python
params={
    'sol':SOL,
    'camera':CAMERA,
    'api_key':apikey
}
```

```python
response=requests.get(URL,params=params)
data=response.json()
photos=data.get('photos',[])
print(f"Found {len(photos)} taken by {CAMERA} on sol {SOL}")
for photo in photos[:10]:
    print(photo['img_src'])
```
