## Serialization

```python
file_name = "Sample.json"
with open(file_name, "w") as fid: 
     json.dump(json_page, fid)
```

## Deserialization

```python
with open(file_name, "r") as fid: 
     json_obj = json.load(fid)

print(json_obj)
```

```python
for key in json_obj:
    print("{:>12}: {}".format(key, json_obj[key]))
```
