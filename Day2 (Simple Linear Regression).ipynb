# Feature Selection/Engg

```python
# False(0) & True(1)
# Another name of "get_dummies" is "One Hot Encoding"

import pandas as pd
import numpy as np
data={'name':['a','b','c'],
     'color':['red','blue','red']}
df=pd.DataFrame(data)
df1=pd.get_dummies(df,columns=['color'])
print(df1)
print()
df2=pd.get_dummies(df,columns=['color'],drop_first=True)
print(df2)
```

```python
# 170
# You've been given a dataset with apartment area and price information. There's a noticeable non-linear relationship between 
# area and price. To address this you intend to categorize them into 'High', 'Medium', and 'Low' groups. Prices above $3,000,000 
# are 'High', below $2,000,000 are 'Low', and between $2,000,000 and $3,000,000 are 'Medium'. Write a code to achieve this 
# assuming that dataset has two columns named area and price.
```

```python
import pandas as pd
import numpy as np
df=pd.DataFrame({
    'area':[1000,1500,2000,5000],
    'price':[1800000,2500000,3200000,2900000]
})
df['category']=np.where(df['price']>3000000,'High',np.where(df['price']<2000000,'Low','Medium'))
print(df)
```

```python
# 171
# In a survey dataset, you have a column representing participants' ages. You want to categorize ages into 'Young', 'Middle-aged',
# and 'Elderly' groups. Ages below 30 are 'Young', ages between 30 and 60 are 'Middle-aged', and ages above 60 are 'Elderly'.
# Write a code to achieve this assuming the dataset has a column named 'age'.
```

```python
import pandas as pd
import numpy as np
df=pd.DataFrame({
    'Name':['a','b','c','d','e'],
    'age':[30,65,28,15,45]
})
df['category']=np.where(df['age']>60,'Elderly',np.where(df['age']<30,'Young','Middle-aged'))
print(df)
```

```python
# 172
# In a customer dataset, you have a column representing customer incomes. You want to categorize incomes into 'Low', 'Medium',
# and 'High' groups. Incomes below 30000 are 'Low', incomes between 30000 and 70000 are 'Medium', and incomes above 70000
# are 'High'. Write a code to achieve this assuming the dataset has a column named 'income'.
```

```python
import pandas as pd
import numpy as np
df=pd.DataFrame({
    'Name':['a','b','c','d','e'],
    'income':[10000,7000,2000,8000,2900]
})
df['category']=np.where(df['income']>7000,'High',np.where(df['income']<3000,'Low','Medium'))
print(df)
```

# Simple Linear Regression

# Template Program

# 1 import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error 
```

# 2.Load Data

```python
df=pd.read_csv('olympic100m.csv')
df
```

# 3. Extract label & feature

```python
x=df[['year']] # Feature
y=df['time']   #Label
```

# 4. Create Model

```python
model=LinearRegression()
```

# 5. Fit Model

```python
model.fit(x,y)
```

# 6. Slop & Intercept

```python
slop=model.coef_[0]
intercept=model.intercept_
```

# 7. Prediction

```python
y_pred=model.predict(x)
print(y_pred)
```

# 8. Evaluation Matrix & Value at 2020

```python
x_2020=np.array([2020]).reshape(-1,1)
model.predict(x_2020)
print(mean_absolute_error(y,y_pred))
print(mean_squared_error(y,y_pred))
print(r2_score(y,y_pred))
```

# 9. Plot

```python
plt.scatter(x,y)
plt.plot(x,y_pred,color='r')
```
